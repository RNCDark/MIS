```python
import pandas as pd
#from sklearn.dummy import DummyClassifier #-for catergorial output
from sklearn.dummy import DummyRegressor #-for quantitativ output
import numpy as np

df = pd.read_csv('data/rpgg_db_1903.csv')
print(df.values)
#get value types
print(df.dtypes)

# Seperate input and output
X = df.loc[:, ['num votes', 'year']]
y = df.loc[:, 'avg rating']

#tune hyper parameter
model = DummyRegressor() 

model.fit(X, y)

unknown = np.array([[496, 2004]])

print(f'Prediction: {model.predict(unknown)}')

# evaluate how good model is doing
model.score(X, y)
```
